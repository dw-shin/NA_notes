Least squares method

```python
import tensorflow as tf

# Model Parameter
W = tf.Variable([.3], dtype=tf.float32)
b = tf.Variable([-.3], dtype=tf.float32)

# Model input and output
x = tf.placeholder(tf.float32)
linear_model = W*x + b
y = tf.placeholder(tf.float32)

# loss
loss = tf.reduce_sum(tf.square(linear_model - y))

# optimizer
optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(loss)

# training data
x_train = [[1],[2],[3],[4]]
y_train = [[0],[-1],[-2],[-3]]
```

```python
MaxIter = 10
# training loop
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    
    for i in range(MaxIter):
        curr_w, curr_b, curr_loss = sess.run([W,b,loss],
                                            feed_dict={x:x_train,
                                                       y:y_train})
        print(i, curr_w, curr_b, curr_loss)
        sess.run(train, feed_dict={x:x_train, y:y_train})
    
    curr_w, curr_b, curr_loss = sess.run([W,b,loss],
                                         feed_dict={x:x_train,
                                                    y:y_train})
    print(i, curr_w, curr_b, curr_loss)
```

```python
import numpy as np

f = lambda x: x**2 - 4*x + 6
x = np.linspace(-1,6,100)
y = f(x)

#Graph
fig = plt.figure(figsize=(15,15))
plt.plot(x,y,'k-')
plt.plot(2,2,'sk')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
grad_f = lambda x: 2*x - 4

x0 = 0.0
MaxIter = 10
learning_rate = 0.25

print('step \t x \t \t f(x)')
print('{:02d} \t {:6.5f} \t {:6.5f}'.format(0,x0,f(x0)))
for i in range(MaxIter):
    x1 = x0 - learning_rate * grad_f(x0)
    x0 = x1
    print('{:02d} \t {:6.5f} \t {:6.5f}'.format(i+1,x0,f(x0)))
```

```python
def steepest_descent(func, grad_func, x0, learning_rate = 0.01,
                    MaxIter=10, verbose=True):
    path=[]
    for i in range(MaxIter):
        x1 = x0 - learning_rate * grad_func(x0)
        if verbose:
            print('{0:03d} \t {1:4.3f} \t {2:3.2f}'.format(i+1,x1,func(x1)))
        x0 = x1
        path.append(x0)
    return (x0,func(x0),np.array(path))
```

```python
xopt, fopt, paths = steepest_descent(f,grad_f, 0.0, learning_rate = .25,
                                    MaxIter = 10)

fig = plt.figure(figsize=(20,10))
plt.plot(x,y,'k-')
plt.plot(2,2,'sk')
plt.plot(paths,f(paths),'ko:')
for k,point in enumerate(paths):
    plt.text(point,f(point), '$x_{0}'.format(k),
            verticalalignment='bottom')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
def steepest_descent_twod(func,gradx,grady,
                         x0,MaxIter=10,
                         learning_rate=0.25):
    paths = [x0]
    fval_paths=[func(x0[0], x0[1])]
    for i in range(MaxIter):
        grad = np.array([gradx(*x0),grady(*x0)])
        x1 = x0 - learning_rate * grad
        fval = func(*x0)
        x0 = x1
        paths.append(x0)
        fval_paths.append(fval)
    paths = np.array(paths)
    paths = np.array(np.matrix(paths).T)
    fval_paths = np.array(fval_paths)
    return(x0,fval,paths,fval_paths)
```

```python
np.random.seed(320)
x_train = np.linspace(-1,1,50)
f = lambda x: 0.5 * x + 1.0
y_train = f(x_train) + 0.4 * np.random.rand(len(x_train))
plt.plot(x_train,y_train,'.k',
         markerfacecolor='none')
plt.grid()
plt.plot()
```

```python
def loss(a,b):
    N = len(x_train)
    val = 0.0
    for i in range(N):
        val += 0.5*(a*x_train[i]+b-
                    y_train[i])**2
    return val/N

def loss_grad_a(a,b):
    N = len(x_train)
    val = 0.0
    for i in range(N):
        er = a * x_train[i] + b - y_train[i]
        val += er * x_train[i]
    return val/N

def loss_grad_b(a,b):
    N = len(x_train)
    val = 0.0
    for i in range(N):
        er = a * x_train[i] + b - y_train[i]
        val += er
    return val/N
```

```python
def visualize_l2(w, b, x_train, y_train, loss):
    from matplotlib.patches import Rectangle
    loss_name = loss.__name__
    plt.plot(x_train,y_train,'.k',
             markerfacecolor='none')
    plt.plot(x_train,w*x_train+b,'--k')
    currentAxis = plt.gca()
    for xx,yy in zip(x_train,y_train):
        currentAxis.add_patch(Rectangle((xx,yy),w*xx+b-yy,w*xx+b-yy, alpha=0.1,facecolor='gray',edgecolor='k'))
    plt.grid()
    plt.axis('equal')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("a={:1.2f}, b={:1.2f} ({}={:5.4f})".format(w,b,loss_name,loss(w,b)))
    
visualize_l2(-4,-1,x_train,y_train,loss)
plt.show()
```

```python
def steepest_descent_twod(func,gradx,grady,
                         x0,MaxIter=10,
                         learning_rate=0.25):
    paths = [x0]
    fval_paths=[func(x0[0], x0[1])]
    for i in range(MaxIter):
        grad = np.array([gradx(*x0),grady(*x0)])
        x1 = x0 - learning_rate * grad
        fval = func(*x0)
        x0 = x1
        paths.append(x0)
        fval_paths.append(fval)
    paths = np.array(paths)
    paths = np.array(np.matrix(paths).T)
    fval_paths = np.array(fval_paths)
    return(x0,fval,paths,fval_paths)
```
